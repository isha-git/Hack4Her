```python
def similar(num1, num2, threshold):
    diff = abs(num1 - num2)
    if diff > threshold:
        return False
    else:
        return True

    
hos_lon = ['4.443347447652479', '4.481427108576028', '4.499108230440157', '4.524170791502297', '4.4975632780416515']
hos_lat = ['51.914242538091266', '51.9102448319788', '51.92565021290264', '51.92390347650823', '51.88657059914318']

matching_lon = []

for lon in hos_lon:
    hos_lon = float(lon)

    for i in range(len(nodes['lon'])):
         if similar(float(hos_lon), float(nodes['lon'][i]), 0.0005) == True:
            matching_lon.append(i)
            print("Hospital lon:" , lon)
            print("Value of i:", i)
            print("Matching lon value:", nodes['lon'][i])
```

```python
hos_lon_datasets = ['4.443347447652479', '4.481427108576028', '4.499108230440157', '4.524170791502297', '4.4975632780416515']
hos_lat_datasets = ['51.914242538091266', '51.9102448319788', '51.92565021290264', '51.92390347650823', '51.88657059914318']


for hos_lon, hos_lat in zip(hos_lon_datasets, hos_lat_datasets):
    matching_indices_lon = []
    matching_indices_lat = []

    hos_lon = float(hos_lon)
    hos_lat = float(hos_lat)

    for i in range(len(nodes['lon'])):
        if similar(hos_lon, float(nodes['lon'][i]), 0.0005) == True:
            matching_indices_lon.append(i)

    for i in range(len(nodes['lat'])):
        if similar(hos_lat, float(nodes['lat'][i]), 0.0005) == True:
            matching_indices_lat.append(i)

    matching_values = []
    
    for i in range(len(matching_indices_lon)):
        for t in range(len(matching_indices_lat)):
            if matching_indices_lon[i] == matching_indices_lat[t]:
                matching_values.append(matching_indices_lon[i])
            
    print("Matching values for hos_lon =", hos_lon, "and hos_lat =", hos_lat, ":", matching_values) #hospital location
```
